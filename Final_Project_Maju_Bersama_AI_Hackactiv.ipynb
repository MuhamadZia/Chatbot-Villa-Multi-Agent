```python
%%writefile base_agent.py
class BaseAgent:

    def format_response(

        self,

        answer,

        sources,

        confidence,

        status

    ):

        return {

            "agent":
            self.name,

            "answer":
            answer,

            "sources":
            sources,

            "confidence":
            confidence,

            "status":
            status

        }
```

```python
%%writefile villa_agent.py

# loader
from langchain_community.document_loaders import DirectoryLoader, TextLoader

loader = DirectoryLoader(
    "./knowledge/villa",
    glob="*.md",
    loader_cls=TextLoader
    )

docs_villa = loader.load()

# chunks
from langchain_text_splitters import MarkdownHeaderTextSplitter

headers = [
    ('#', 'Header1'),
    ('##', 'Header2')
]

splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers,
    strip_headers=False
)

chunck_all = []

for doc in docs_villa:
  chunck = splitter.split_text(doc.page_content)
  for c in chunck:
    c.metadata = {
        **doc.metadata,
        **c.metadata
    }
  chunck_all.extend(chunck)

# print(len(chunck_all))

# embedding
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embedding_model = GoogleGenerativeAIEmbeddings(
    # api_key=gemini_api_key,
    model="gemini-embedding-2",
    output_dimensionality=768
)

# store
from langchain_chroma import Chroma

vectordb = Chroma.from_documents(
    documents=chunck_all,
    embedding=embedding_model,
    persist_directory='./villa_db'
)

retriever = vectordb.as_retriever(search_kwargs={"k":2})

# agent class
from base_agent import BaseAgent
class VillaAgent(BaseAgent):

    def __init__(
        self,
        retriever,
        llm
    ):
        self.retriever = retriever
        self.llm = llm
        self.name = "villa"

    def ask(
        self,
        question
    ):

        # mendapat doc yang relevant
        docs = self.retriever.invoke(
          question
        )

        # sources
        sources = []

        for d in docs:
          sources.append(
              d.metadata['source']
          )

        sources = list(set(sources))

        # context
        context = "\n\n".join(
          d.page_content
          for d in docs
        )

        system_prompt = f"""
        Anda adalah Agent Villa NR.

        Tugas Anda:

        - Membantu calon tamu.
        - Gunakan context yang diberikan.
        - Jangan mengarang informasi.
        - Jika informasi tidak tersedia,
          katakan Anda belum memiliki informasi tersebut.
        - Jangan bertanya.
        - Jangan menyapa.

        Context:

        {context}

        Pertanyaan:

        {question}
        """

        response = self.llm.invoke(system_prompt)

        return self.format_response(
          answer = response.content,
          sources = sources,
          confidence = "high",
          status = "success"

      )
```

```python
%%writefile tools_agent.py

from langchain_core.tools import tool
from villa_agent import retriever, VillaAgent
from travel_agent import retriever_travel, TravelAgent
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0,
)

@tool
def search_villa_info(query:str) -> str:
  """
  Use this tool to answer question of user for question related to villa nature information.
  Do not ask follow-up questions.
  Return ONLY relevant facts.
  """
  # """
  # Use this tool to retrieve relevant of villa nature information.

  # Return ONLY relevant facts.

  # Do not greet.

  # Do not ask follow-up questions.

  # Do not recommend unrelated information.

  # If no relevant information exists:

  # Return:

  # NO_RELEVANT_INFORMATION
  # """
  # """
  #   Use this to search factual information about Villa Nature.

  #   Use ONLY for:
  #   - rooms
  #   - facilities
  #   - pricing
  #   - booking
  #   - policies
  #   - activities inside Villa NR
  #   - promotions
  #   - food
  #   - FAQs
  # """
  global llm
  villa_agent = VillaAgent(retriever=retriever,llm=llm)
  return villa_agent.ask(query)['answer']

@tool
def search_travel_info(query:str) -> str:
  """
  Use this tool to answer question of user for question related to travel to villa nature.
  Do not ask follow-up questions.
  Return ONLY relevant facts.
  """
  # """
  # Use this tool to retrieve relevant of travel to villa nature information.

  # Return ONLY relevant facts.

  # Do not greet.

  # Do not ask follow-up questions.

  # Do not recommend unrelated information.

  # If no relevant information exists:

  # Return:

  # NO_RELEVANT_INFORMATION
  # """
  # """
  #   Use this to search travel information around Villa Nature.

  #   Use ONLY for:
  #   - attractions
  #   - transportation
  #   - itineraries
  #   - restaurants
  #   - nearby destinations

  # """
  travel_agent = TravelAgent(retriever=retriever_travel,llm=llm)
  return travel_agent.ask(query)['answer']

assistant_tools = [search_villa_info, search_travel_info]
```

```python
%%writefile assistant_agent.py
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
from tools_agent import assistant_tools,llm
# from langchain_groq import ChatGroq
# from google.colab import userdata

system_instruction = """
You are Villa NR Master Assistant.

Your primary responsibility is to help users by providing accurate, friendly, and natural information about Villa NR and travel experiences related to Villa NR.

You have access to specialist tools. Use them whenever additional information is needed.

Always act as a single assistant. Never mention specialist agents, tools, internal reasoning, or internal workflows to the user.

---

# YOUR SCOPE

You are ONLY able to help with:
1. Information about Villa Nature (facilities, rooms, prices, rules, etc.)
2. Travel information related to Villa Nature (routes, transportation, nearby places, etc.)
3. General conversation directly related to the two topics above.

You are NOT a general assistant.
You CANNOT help with tasks, questions, or topics unrelated to Villa Nature and travel to Villa Nature.

---

# OUT-OF-SCOPE REQUESTS

If the user asks anything outside your scope — including general tasks, unrelated questions, or requests for physical help — respond politely but firmly.

Do NOT attempt to answer or fulfill out-of-scope requests.
Do NOT say "I can help with that" for anything outside your scope.

Use responses like:
- "Maaf, saya hanya bisa membantu informasi seputar Villa Nature dan perjalanan ke Villa Nature."
- "Pertanyaan itu di luar cakupan saya. Ada yang ingin ditanyakan tentang Villa Nature?"
- "Saya adalah asisten khusus Villa Nature, jadi saya tidak bisa membantu hal tersebut. Apakah ada yang ingin Anda ketahui tentang villa atau perjalanan ke sini?"

Examples of out-of-scope requests and how to respond:

User:
Bisa bantu angkat galon?

A:
Maaf, itu di luar kemampuan saya. Saya hanya bisa membantu informasi seputar Villa Nature dan perjalanan ke sana. Ada yang ingin Anda tanyakan?

User:
Tolong buatkan kode Python untuk saya.

A:
Maaf, saya hanya bisa membantu informasi seputar Villa Nature dan perjalanan ke Villa Nature. Ada yang ingin Anda ketahui tentang villa kami?

User:
Siapa presiden Indonesia?

A:
Pertanyaan itu di luar cakupan saya. Saya adalah asisten khusus Villa Nature — ada yang ingin ditanyakan tentang villa atau perjalanan ke sini?

---

# AVAILABLE TOOLS

## search_villa_info

Always use this tool if the question requires factual Villa Nature information.

## search_travel_info

Always use this tool if travel knowledge to Villa Nature and its location.

# TOOL RESULTS ARE AUTHORITATIVE.

Information returned from specialist tools is the primary source of truth.

After receiving tool results:

- Use only information returned by the tools.
- Do not add assumptions.
- Do not invent facilities.
- Do not infer prices.
- Do not add common knowledge.
- Do not expand beyond retrieved facts.

If a fact is not present in tool outputs, do not mention it.

---

# RESPONSE GENERATION

After receiving information from the necessary tools:

- Combine all relevant information into one natural response.
- Remove duplicate information.
- Organize the answer logically.
- Prioritize clarity and helpfulness.
- Make the response conversational and friendly.
- Always generate a final response after tool execution.

Never expose raw tool outputs.

Never say:
"I called a tool."
"The Villa Agent says."
"The Travel Agent says."

Present everything as one seamless answer.

---

# TOOL USAGE

Decide which specialist tools are required.

- If the question is only about Villa Nature, use search_villa_info.
- If the question is only about travel, use search_travel_info.
- If the question requires both Villa Nature and travel knowledge, use both tools.
- You may call multiple tools before producing a final answer.

Prefer specialist tools over your own knowledge whenever relevant.

---

# GENERAL CONVERSATION

For greetings, casual conversation, gratitude, and small talk
that are still within the context of Villa Nature:

- Respond naturally.
- Do not call specialist tools.
- Keep the response short.
- Politely redirect to Villa Nature topics if needed.

Examples:

User:
Halo

A:
Halo! Senang bertemu dengan Anda. Ada yang bisa saya bantu mengenai Villa Nature atau perjalanan Anda?

User:
Apa kabar?

A:
Kabar baik, terima kasih. Ada yang bisa saya bantu terkait Villa Nature atau rencana perjalanan Anda?

User:
Terima kasih

A:
Sama-sama! Senang bisa membantu. Jika ada pertanyaan tentang Villa Nature atau perjalanan Anda, silakan beri tahu saya.

"""

# Create a checkpointer for memory
checkpointer = InMemorySaver()

# Create the ReAct agent
# llm = ChatGroq(
#     model="llama-3.3-70b-versatile",
#     temperature=0,
# )

assitant_agent = create_agent(
    model=llm,
    tools=assistant_tools,
    system_prompt=system_instruction,
    checkpointer=checkpointer
)

# create chat
def chat(user_message, thread_id):

    config = {
        "configurable": {
            "thread_id": thread_id
        }
    }

    response = assitant_agent.invoke(
        {
            "messages": [
                {
                    "role":"user",
                    "content":user_message
                }
            ]
        },
        config=config
    )

    # uncomment for debugging
    for i,m in enumerate(response["messages"]):
      print("="*50)
      print(type(m))
      print(m)

    return response["messages"][-1].content
```

```python
%%writefile streamlit_chat_app.py

import streamlit as st
from assistant_agent import chat  # sesuaikan import dengan lokasi fungsi chat()

# =====================================================
# Title
# =====================================================

st.title("Villa NR Chatbot")
st.caption("Chatbot menggunakan Villa NR Agent")

# =====================================================
# Sidebar
# =====================================================

with st.sidebar:

    st.subheader("Pengaturan")

    reset_button = st.button("Reset Percakapan")

# =====================================================
# Session State Init
# =====================================================

if "messages" not in st.session_state:
    st.session_state.messages = []

if "thread_id" not in st.session_state:
    st.session_state.thread_id = "default-thread"  # atau pakai uuid

# =====================================================
# Reset
# =====================================================

if reset_button:
    st.session_state.messages = []
    st.session_state.thread_id = "default-thread"  # reset thread juga
    st.rerun()

# =====================================================
# Render History
# =====================================================

for msg in st.session_state.messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

# =====================================================
# User Input
# =====================================================

prompt = st.chat_input("Ketik pesan...")

if prompt:

    # User

    st.session_state.messages.append(
        {
            "role": "user",
            "content": prompt
        }
    )

    with st.chat_message("user"):
        st.markdown(prompt)

    # Assistant

    try:

        with st.chat_message("assistant"):
            with st.spinner("Sedang berpikir..."):

                answer = chat(
                    user_message=prompt,
                    thread_id=st.session_state.thread_id
                )

                st.markdown(answer)

    except Exception as e:

        answer = f"Error: {e}"

        with st.chat_message("assistant"):
            st.error(answer)

    st.session_state.messages.append(
        {
            "role": "assistant",
            "content": answer
        }
    )
```

```python
import subprocess
import time

def run_streamlit(filename, port=8501):
    # Kill SEMUA proses streamlit, bukan hanya yang kita spawn
    subprocess.run(["pkill", "-f", "streamlit"], capture_output=True)

    # Force-free port kalau masih ada yang nempel
    subprocess.run(["fuser", "-k", f"{port}/tcp"], capture_output=True)

    # Tutup semua tunnel ngrok
    ngrok.kill()

    # Tunggu port benar-benar bebas
    time.sleep(3)

    proc = subprocess.Popen(
        [
            "streamlit", "run", filename,
            "--server.headless=true",
            "--server.port", str(port),
            "--server.enableCORS=false",
        ],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL
    )

    time.sleep(3)

    public_url = ngrok.connect(port)
    print(f"Streamlit berjalan: {public_url}")

    return proc
```
